```python
import wrds
db = wrds.Connection()
index_div_yield = db.raw_sql("""
                                select  b.issue_type, 
                                        b.index_flag, 
                                        b.exchange_d, 
                                        b.sic, 
                                        b.secid, 
                                        b.ticker, 
                                        b.cusip, 
                                        a.date, 
                                        a.rate  
                                from optionm_all.idxdvd a, optionm_all.securd b
                                where a.secid = b.secid
                                and b.ticker = 'SPX'
                                and a.date >= '01/01/2020'
                                and a.date <= '12/31/2021'
                                order by a.date
                             """, 
                             date_cols=['a.date'])
index_div_yield.shape
```
