```python
apple_2007_0109 = db.raw_sql("""
                                select  b.secid,
                                        a.date,
                                        a.volatility,
                                        b.cusip,
                                        b.ticker
                                 from optionm_all.hvold2007 a, optionm_all.securd1 b
                                where a.secid = b.secid
                                  and b.ticker = 'AAPL'
                                  and a.date >= '01/01/2007'
                                  and a.date <= '01/23/2007'
                                  and a.days = '10'
                                  --order by a.date
                                UNION
                                select  d.secid,
                                        c.date,
                                        c.volatility,
                                        d.cusip,
                                        d.ticker
                                 from optionm_all.hvold2006 c, optionm_all.securd1 d
                                where c.secid = d.secid
                                  and d.ticker = 'AAPL'
                                  and c.date >= '12/26/2006'
                                  and c.date <= '12/31/2006'
                                  and c.days = '10'
                                  --order by b.date
                                 order by 1
                             """,
                             date_cols=['a.date', 'c.date'])
apple_2007_0109
```
