```python
amazon_ibm = db.raw_sql("""select datadate, gvkey, conm, tic, cusip, atq, ltq, niq, teqq
                        from comp_na_daily_all.fundq
                        where (tic = 'IBM' or tic = 'AMZN')
                        and datadate>='01/01/2010' and datadate<='12/31/2020'""",
                        date_cols=['datadate'])
amazon_ibm
```
