```python
import wrds
db = wrds.Connection()
```

```python
lib = db.list_libraries()
lib
```

```python
tables = db.list_tables(library='tr_worldscope')
tables
```

```python
db.describe_table(library='tr_worldscope', table='wsndata')
```

```python
# 03151 Working Capital 
item_03151 = db.raw_sql("""
                        select count (*) 
                          from tr_worldscope.wsndata
                         where item = 03151
                        """)
item_03151
```

```python
# 03151 Working Capital 
item_03151 = db.raw_sql("""
                        select *--count (*) 
                          from tr_worldscope.wsndata
                         where item = 03151
                         limit 10
                        """)
item_03151
```

```python
# 03151 Working Capital 
item_03151 = db.raw_sql("""
                        select *--count (*) 
                          from tr_worldscope.wsndata
                         where item = 03151
                           and freq <> '@'
                         limit 10
                        """)
item_03151
```

```python
# 03151 Working Capital 
item_03151 = db.raw_sql("""
                        select *--count (*) 
                          from tr_worldscope.wsndata
                         where item = 03151
                           and freq <> '@'
                           and year_ = 2019
                         limit 10
                        """)
item_03151
```

```python
item_03151.to_csv('tr_worldscope__wsndata_03151.csv')
```

```python
# 03151 Working Capital 
item_03151 = db.raw_sql("""
                        select *--count (*) 
                          from tr_worldscope.wsndata
                         where item = 03151
                           and freq <> '@'
                           and year_ >= '2019'
                           and year_ <= '2023'
                         --and year_ in ('2019', '2020', '2021', '2022', '2023')
                         limit 10
                        """)
item_03151
```

```python
item_03151 = db.raw_sql("""
                        select * 
                          from tr_worldscope.wsitem
                         where number = 03151
                        """)
item_03151
```

```python
table_to_use = db.raw_sql("""
                        select table_
                          from tr_worldscope.wsitem
                      group by table_
                      order by 1
                        """)
table_to_use
```
