```python
import wrds
db = wrds.Connection()
```

```python
lib = db.list_libraries()
lib
```

```python
tables = db.list_tables(library='tr_worldscope')
tables
```

```python
db.describe_table(library='tr_worldscope', table='wsndata')
```

```python
# 18157 Capital Adequacy Ratio - Tier 1 
item_18157 = db.raw_sql("""
                        select count (*) 
                          from tr_worldscope.wsndata
                         where item = 18157
                        """)
item_18157
```

```python
# 18157 Capital Adequacy Ratio - Tier 1 
item_18157 = db.raw_sql("""
                        select *--count (*) 
                          from tr_worldscope.wsndata
                         where item = 18157
                         limit 10
                        """)
item_18157
```

```python
item_18157.to_csv('tr_worldscope__wsndata_18157.csv')
```

```python
# 15061 Non-Performing Loans % Total Loans
item_15061 = db.raw_sql("""
                        select *--count (*) 
                          from tr_worldscope.wsndata
                         where item = 15061
                           and year_ >= '2019'
                           and year_ <= '2023'
                         --and year_ in ('2019', '2020', '2021', '2022', '2023')
                         limit 10
                        """)
item_15061
```
