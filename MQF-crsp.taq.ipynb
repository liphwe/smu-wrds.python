```python
import wrds
db = wrds.Connection()
taq = db.raw_sql("""
                    select 
                    case when sym_suffix is not null then 'CHK' end as "CHK", a.*
                    from taqm_2020.cqm_20200102 a
                    where sym_root = 'AAPL'
                    --and date ='01/02/2020'
                    order by 1 --time_m
                """)
taq
```
