# Question 7: How do I query single stock 'Apple' from daily stock file of CRSP?

---


# Step 7: Subsetting Data for single stock 'AAPL'

---



*   raw_sql()method
*   when "conditioning" is needed
familiar with SQL syntax
*   can pre-specify date column format
*   Example: Condition for single ticker AAPL (under permno = 14593) and setting the date between 2010 to 2020 to run time series for data variables: daily stock price, open price and ask price.
* apple is the DataFrame name I use 




subsetting data Apple.png

```python
# Select single stock's daily price (prc), open price (openprc) and ask price (ask) 
# Select single stock AAPL, identifier to use permno = 14593
# between 2010 and 2020
# apple is a dataframe that I created

apple = db.raw_sql("""select permno, date, cusip, vol, permco, openprc, prc, ask
                        from crsp.dsf 
                        where permno = 14593
                        and date>='01/01/2010' and date<='12/31/2020'""", 
                     date_cols=['date'])
apple
```

# Step 8: Subsetting Data for two stocks 'Apple and IBM'

*   raw_sql()method
*   when "conditioning" is needed
familiar with SQL syntax
*   can pre-specify date column format
*   Example: Condition for two stocks IBM (under permno = 12490) and Apple (under permno = 14593) and setting the date between 2010 to 2020 to run time series for data variables: daily stock price, open price and ask price.


subsetting two stocks Apple and IBM.png

```python
# Select two stocks' daily price
# E.g. IBM, permno = 12490 and Apple, permno = 14593
# Date range between 2010 to 2020 onwards
# ibm_apple_dsf is a name that I created

ibm_apple_dsf = db.raw_sql("""select permno, date, cusip, vol, permco, openprc, prc, ask
                        from crsp.dsf 
                        where (permno = 12490 or permno = 14593) 
                        and date>='01/01/2010' and date<='12/31/2020'""", 
                     date_cols=['date'])
ibm_apple_dsf
```

## Advanced - Joining or Mergeting 2 Tables

```python
db.raw_sql("""
    SELECT a.gvkey, a.datadate, a.tic, a.conm, a.at, a.lt, b.prccm, b.cshoq
    FROM comp.funda a
    JOIN comp.secm b ON a.gvkey = b.gvkey AND a.datadate = b.datadate
    WHERE a.tic = 'IBM' AND a.datafmt = 'STD' AND a.consol = 'C' AND a.indfmt = 'INDL'
""")
```
